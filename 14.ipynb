```python
from numba import jit
import matplotlib.pyplot as plt
import numpy as np
import time

n = [10, 100, 1000, 10000, 100000, 1000000]

def oszac_pi():
    for x in n:
        start_time = time.time()
        a = np.random.rand(x)
        b = np.random.rand(x)
        area = 15
        colors = np.where(a**2+b**2 <= 1, 'g', 'r')
        plt.scatter(a, b, s=area, c=colors)
        ile_ziel = list(colors)
        oszac = ile_ziel.count('g')
        plt.show()
        oszac_pi = 4*oszac/(x)
        end = time.time()
        print('Pi oszacowane przy', x, 'punktach to', oszac_pi)
        print('%s sekund' % (time.time() - start_time))
    
@jit(forceobj=True)
def numba_pi():
    for x in n:
        start_time = time.time()
        a = np.random.rand(x)
        b = np.random.rand(x)
        area = 15
        colors = np.where(a**2+b**2 <= 1, 'g', 'r')
        plt.scatter(a, b, s=area, c=colors)
        ile_ziel = list(colors)
        oszac = ile_ziel.count('g')
        plt.show()
        oszac_pi = 4*oszac/(x)
        end = time.time()
        print('Pi oszacowane przy', x, 'punktach to', oszac_pi)
        print('%s sekund' % (time.time() - start_time))

oszac_pi()
numba_pi()
```
