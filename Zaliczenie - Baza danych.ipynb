```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame

# lf - prepared earlier version, can be replaced by using option 6 when running baza() to create empty database
lf = pd.DataFrame({
    'Title':['Fahrenheit 451', 'Harry Potter and the Chamber of Secrets', 'Silmarillion', 'The Way of Kings', 'Twilight',
             'Dune', 'A Brief History of Time', 'The Selfish Gene', 'To Kill a Mockingbird', 'The Stand'],
    'Author':['Ray Bradbury', 'Joanne Rowling', 'John Tolkien', 'Brandon Sanderson', 'Stephanie Meyer',
             'Frank Herbert', 'Stephen Hawking', 'Richard Dawkins', 'Harper Lee', 'Stephen King'],
    'Published':['1953', '1998', '1977', '2010', '2005',
                '1976', '1960', '1988', '1965', '1978'],
    'Genre':['Dystopic', 'Fantasy', 'Fantasy', 'Fantasy', 'Fantasy',
               'Science fiction', 'Popular science', 'Popular science', 'Bildungsroman', 'Post-apocalyptic']})
lf.to_csv('Bibliografia.csv', index=False)
```

```python
# Work pipeline - each modification creates a new file. As such functions run inside baza() need to load 
# the CSV into DataFrame every time, otherwise displayed DataFrame will be outdated and unmodified.
# Option 1 loads and displays the DataFrame. The file name is hardcoded - unfortunately, time constraints didn't allow me to create a relative naming/pathing scheme.
# File is modified by options 2-5, using inputs from the use, converted to string or integer, as suitable. Afterwards,
# the file is exported to CSV again. In case of no CSV file being present, option 6 will create an empty one. 
# Option 7 will quit the program. baza() is setup so that after executing options 1-6, user will be redirected back to the menu, allowing for chosing other option, or quitting.
```

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame

# Creates second dataframe, and then concatenates it to the original one. 
# Uses inputs from user to prepare new entry.
def dodaj():
    library = pd.read_csv('Bibliografia.csv')
    a = input('Title?')
    b = input('Author?')    
    c = input('Year published?')
    d = input('Genre?')
    nowy = pd.DataFrame({'Title': [a], 'Author': [b], 'Published': [c], 'Genre': [d]})
    library = pd.concat([library, nowy], ignore_index=True)
    library.to_csv('Bibliografia.csv', index=False)

# Modifies specific value in a chosen column.
# Column chosen must be typed in by the user exactly as shown, otherwise error will be raised.
def zmien():
    library = pd.read_csv('Bibliografia.csv')
    no = int(input('Which entry would you like to change? Type the number in front of the book title you\'d like to modify.\n'))
    part = str(input('Which field would you like to change? Type one of the options below exactly as shown.\nTitle\nAuthor\nYear published\nGenre\n'))
    if part == 'Title':
        pass
    elif part == 'Author':
        pass
    elif part == 'Year published':
        pass
    elif part == 'Genre':
        pass
    else:
        raise ValueError('No field found under name typed.')
    new = str(input('New description?\n'))
    library.at[no, part] = new
    library.to_csv('Bibliografia.csv', index=False)
    
# Replaces whole row with new values. New values come from user inputs.
def zamien():
    library = pd.read_csv('Bibliografia.csv')
    no = int(input('Which entry would you like to change? Type the number in front of the book title you\'d like to modify.\n'))
    a = str(input('New title?'))
    b = str(input('New author?'))
    c = str(input('New year published?'))
    d = str(input('New genre?'))
    library.loc[no, ['Title','Author', 'Published', 'Genre']] = [a, b, c, d]
    library.to_csv('bibliografia.csv', index=False)

# Removes specific row from database, as chosen by the user.
def usun():
    library = pd.read_csv('Bibliografia.csv')
    display(library)
    no = int(input('Which entry would you like to delete?\n'))
    library = library.drop([no])
    library.to_csv('Bibliografia.csv', index=False)
    
# Function encompassing the interactive layer. Has an option to quit, and running other 
# options will lead user back to main menu after function finishes its tasks.
# Option 1 displays the library, options 2-5 modify the contents of the database.
# Option 6 creates new empty database. Option 7 quits the program.
def baza():
    print('Book database - choose an option:\n1. Display book entries\n2. Add book entry\n3. Modify part of book entry\n4. Replace whole book entry\n5. Remove entry\n6. Create new empty database\n7. Quit')
    opcja = int(input())

    if opcja == 1:
        library = pd.read_csv('Bibliografia.csv')
        display(library)
        input('Press any key to go back to menu\n')
        baza()
        
    if opcja == 2:
        dodaj()
        library = pd.read_csv('Bibliografia.csv')
        display(library)
        input('Press any key to go back to menu\n')
        baza()
        
    if opcja == 3:
        zmien()
        library = pd.read_csv('Bibliografia.csv')
        display(library)
        input('Press any key to go back to menu\n')
        baza()

    if opcja == 4:
        zamien()
        library = pd.read_csv('Bibliografia.csv')
        display(library)
        input('Press any key to go back to menu\n')
        baza()
        
    if opcja == 5:    
        usun()
        library = pd.read_csv('Bibliografia.csv')
        display(library)
        input('Press any key to go back to menu\n')
        baza()
        
    if opcja == 6:
        library = pd.DataFrame(list())
        library.to_csv('Bibliografia.csv')
        pd.read_csv('Bibliografia.csv')
        display(library)
        input('Press any key to go back to menu\n')
        baza()
        
    if opcja == 7:
        quit()
```
