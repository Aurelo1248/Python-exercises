```python
# Plot functions sin(x), cos(x), and exp(-x) in a range [0,10]. 
# Colors are red, green, and blue, respectively. 
# Lines are solid, dashed, and dotted, respectively. Add a legend.

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x)
line1, = plt.plot(x, y1, color=(1,0,0), linestyle='solid')
line2, = plt.plot(x, y2, color=(0,1,0), linestyle='dashed')
line3, = plt.plot(x, y3, color=(0,0,1), linestyle='dotted')
plt.legend((line1, line2, line3), ('sin(x)', 'cos(x)', 'exp(-x)'), loc=(1.04,0))
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(0,10)
plt.ylim(-1,1)
plt.grid(color='grey', linestyle='-', linewidth=0.3, alpha=0.4)
plt.show()
```
